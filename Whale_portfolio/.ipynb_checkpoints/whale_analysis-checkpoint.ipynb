## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
tsx_60_returns = pd.read_csv("sp_tsx_history.csv", index_col = "Date", parse_dates=True)
```

```python
# Calculate Daily Returns
tsx_daily_returns = tsx_60_returns.pct_change().dropna()
```

```python
tsx_daily_returns
```

```python
# Rename `Close` Column to be specific to this portfolio.
tsx_daily_returns = tsx_daily_returns.rename(columns={'Close': 'TSX_60'})
```

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
combined_data_rolling_120 = combined_data.rolling(window=21).std()

# Plot the rolling standard deviation
sns.set()
sns.set_style("whitegrid")
fig, ax= plt.subplots(figsize=(20,20))
sns.lineplot(ax=ax, data=combined_data_rolling_120, linewidth = 2.5)
plt.legend(labels= [x for x in combined_data.columns], fontsize = 20)
```

```python
combined_data_rolling_120.dropna()
```

### Calculate and plot the correlation

```python
# Calculate the correlation
portfolios_corr = combined_data.corr()

# Display de correlation matrix
sns.set()
sns.set_style("whitegrid")
mask = np.triu(np.ones_like(portfolios_corr, dtype=bool))
fig, ax= plt.subplots(figsize=(20,20))
cmap = sns.color_palette("mako", as_cmap=True)
sns.heatmap(ax=ax, data=portfolios_corr, cmap=cmap, mask=mask, vmax=1, vmin=-1, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
berkshire_cov = combined_data.cov()['BERKSHIRE HATHAWAY INC'].mean()

# Calculate variance of S&P TSX
tsx_var = combined_data['TSX_60'].var()
# Computing beta
res = stats.linregress(combined_data['TSX_60'].values,combined_data['BERKSHIRE HATHAWAY INC'].values)
beta = res.slope
# Plot beta trend
rolling_covariance = combined_data['BERKSHIRE HATHAWAY INC'].rolling(window=30).cov(combined_data['TSX_60'])
rolling_variance = combined_data['TSX_60'].rolling(window=30).var()
rolling_beta = rolling_covariance / rolling_variance
sns.set(font_scale = 2)
sns.set_style("whitegrid")
fig, ax= plt.subplots(figsize=(20,20))
p = sns.lineplot(ax=ax, data=rolling_beta,linewidth = 2.5).set_title('30 Day Rolling Beta')
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
combined_data_ewma = combined_data.ewm(span=21).mean()
```

```python
combined_data_ewma
```

```python
sns.set(font_scale = 2)
sns.set_style("whitegrid")
fig, ax= plt.subplots(figsize=(20,20))
sns.lineplot(ax=ax, data=combined_data_ewma, linewidth = 2.5)
plt.legend(labels= [x for x in combined_data.columns], fontsize = 20)
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
#Assuming risk free rate of 2%
sharpe_ratios = ((combined_data.mean() * 252) - 0.02) / (combined_data.std() * np.sqrt(252))
sharpe_ratios
```

```python
sharpe_df = pd.DataFrame(sharpe_ratios).reset_index().rename(columns={'index':'Portfolio', 0:'Sharpe_Ratio'})
```

```python
# Visualize the sharpe ratios as a bar plot
sns.set(font_scale = 2)
sns.set_style("whitegrid")
fig, ax= plt.subplots(figsize=(20,20))
sns.barplot(ax=ax, x='Portfolio', y='Sharpe_Ratio', data=sharpe_df)
plt.legend(labels= [x for x in combined_data.columns], fontsize = 20)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
```
